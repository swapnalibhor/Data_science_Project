# Dynamic URL

```python
from flask import Flask

my_app = Flask(__name__)  ### Wrenkzug Web App 

@my_app.route('/') ### API
def one():
    return "API One"


@my_app.route('/admin/<name>') ### API
def admin(name):
    return "name is URL ="+name


@my_app.route('/value/<no>') ### API
def value(no):
    return "name is URL ="+no


@my_app.route('/value1/<int:no>') ### API
def valuee(no):
    return "name is URL ="+no


@my_app.route('/value2/<int:no>') ### API
def valueee(no):
    return "name is URL ="+str(no)


@my_app.route('/value3/<float:no>') ### API
def valueeee(no):
    return "name is URL ="+no


@my_app.route('/value4/<float:no>') ### API
def valueeeee(no):
    return "name is URL ="+str(no)

if __name__ == '__main__':
    my_app.run(host= '127.0.0.1', port = 5000, debug= True)


    ################# Basic Web app / flask app is ready to execute 
```

```python
from flask import Flask,redirect,url_for

app = Flask(__name__)

@app.route('/')
def index():
    return "INDEX API"

@app.route('/fail/<int:score>')
def fail(score):
    return "The Person is failed and mark is = " + str(score)


@app.route('/success/<int:score>')   ### URL 
def two(score):      ### Function declartion 
    return "The Person is Passed and mark is = " + str(score)

##### RESULT CHECKING 
@app.route('/user/<int:mark>')
def result_check(mark):
    if mark >= 35 :
        result = 'success'
    else:
        result = 'fail'
    return redirect(url_for(result,score=mark))   #### URL Redirection 

if __name__ == '__main__':
    app.run(host='127.0.0.1',port = 5000, debug = True)



    ######################

    # app.route(url, options)

    # url = '/' or '/admin' or '/predict' or '/result/pass'
    # options ==  rule to set for url 

```

```python
from flask import Flask, redirect, url_for

app = Flask(__name__)

@app.route('/')   # API
def index():
    return "API Point"

@app.route('/admin/<user>')
def one(user):
    return "Welcome admin " + user 

@app.route('/guest/<user>')
def two(user):
    return "Welcome Guest " + user 


@app.route('/userdata/<name>')
def three(name):
    if name == 'abhijit':
        result = 'admin'
    else:
        result = 'guest'
    return redirect (url_for(result,user=name))


if __name__ =='__main__':
    app.run(host='127.0.0.1',port=5000, debug=True)
```
